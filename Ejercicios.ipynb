```
class Punto:
    def __init__(self,x,y):
        self.x = x
        self.y = y

    def eje_x(self):
        return self.x

    def eje_y(self):
        return self.y

    def impresion(self):
        return f"({self.x}, {self.y})"

    def opuesto(self):

        return Punto(-self.x, -self.y)

    def positivo_o_negativo(self):
        if self.x >= 0 and self.y >= 0:
            return "Ambos positivos"
        else:
            self.x < 0 and self.y < 0
            return "Ambos negativos"

p = Punto(2, 5)
p2 = p.opuesto()
print(p.impresion())
print(p2.impresion())
```
